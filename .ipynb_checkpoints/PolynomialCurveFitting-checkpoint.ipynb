## Explain overfitting and underfitting with polynomial model.


```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1)
num_of_points = 9 # Number of training examples

x_train = np.linspace(0, 1, num_of_points)
y_train = np.sin(2*np.pi*x_train) 
y_train += np.random.normal(scale=0.2, size=num_of_points) # ADD Gaussian Noise, scale controls scale of noise

num_of_points = 100 # Number of testing examples 
x_test = np.linspace(0, 1, num_of_points)
y_test = np.sin(2*np.pi*x_test) 
```

```python
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, c="g", label="$\sin(2\pi x)$")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (20,10))
for i, value in enumerate([3,9]):
    plt.subplot(1,2,i+1)
    plt.scatter(x_train, y_train, marker = 'x', label = 'Training data')
    
# for training data    
    x_t = x_train[:, np.newaxis]  # add new dimensions for x^2, x^3,...
    x_ploy= PolynomialFeatures(degree = value).fit_transform(x_t)
    
    model = LinearRegression()
    model.fit(x_ploy, y_train)
    y_train_predict = model.predict(x_ploy)
    
    rmse = np.sqrt(mean_squared_error(y_train_predict, y_train))
    r2 = r2_score(y_train_predict, y_train)
    print("==================training data================")
    print(f"Degree of Polynomial = {value}, rmse = {rmse}, r2 = {r2}")
    
    plt.scatter(x_train, y_train_predict, s=20, label = 'Predicted Result')

    

#For testing data    
    x_t = x_test[:, np.newaxis]  # add new dimensions for x^2, x^3,... For testing data
    x_ploy = PolynomialFeatures(degree = value).fit_transform(x_t)
    y_test_predict = model.predict(x_ploy)
    
    rmse = np.sqrt(mean_squared_error(y_test_predict, y_test))
    r2 = r2_score(y_test_predict, y_test)
    print("==================testing data==================")
    print(f"Degree of Polynomial = {value}, rmse = {rmse}, r2 = {r2}")
    
    plt.plot(x_test, y_test_predict, label = 'Testing Data')
    plt.title( "Degree of Polynominal = " + str(value))
    plt.ylim((-1.5,1.5))
    
plt.legend(bbox_to_anchor=(1.05, 0.64), loc=2, borderaxespad=0.)
plt.show()
```

### The model with degree of ploynominal 9 performs "perfect" on training data, rmse is almost 0 and $r^2$ is 1, but performs worse in testing data => overfitting
